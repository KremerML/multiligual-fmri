```python
model = BertModel.from_pretrained('bert-base-uncased')
```

```python
import ast

def str_to_numpy_array(s):
    return np.array(ast.literal_eval(s))

def get_hidden_representations(embeddings):
    # Convert embeddings to a tensor
    bert_embeddings_tensor = torch.tensor(np.stack(embeddings))

    # Get the hidden representations from the BERT model
    with torch.no_grad():
        hidden_representations, _ = BertModel(bert_embeddings_tensor)

    return hidden_representations.squeeze(0)
```

```python
BERTEmb['BERT'] = BERTEmb['BERT'].apply(str_to_numpy_array)
BERTEmb['hidden_representations'] = get_hidden_representations(BERTEmb['BERT'].values)
```
